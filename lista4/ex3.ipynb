```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = mpimg.imread('stinkbug.png') #carregar imagem de teste
plt.imshow(img) 
```

```python
def suave1(img):
    img_new = np.copy(img) #cria copia desvinculada da original
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            soma = np.copy(img[i][j]) # eu tava passando o ponteiro e o valor de soma estava ligado ao array originial --->  dava tudo errado
            cont = 1                  # ambiguidade em atribuir valor de img[i][j] a soma ou atribuir ponteiro a variavel
            
            #em resumo, faz a mdia dos valores de cada pixel
            #soma  um array de dim 3

            if i >=1:  #verificacao de fronteira
                soma += img[i-1][j] 
                cont += 1
            if i <= (img.shape[0] -2): #verificacao de fronteira
                soma += img[i+1][j] 
                cont += 1
                
            if j >=1: #verificacao de fronteira
                soma += img[i][j-1] 
                cont += 1
            if j <= (img.shape[1] -2): #verificacao de fronteira
                soma += img[i][j+1] 
                cont += 1
    
            img_new[i][j] = soma/cont #atualiza valor suavizado na copia
            
    return img_new
    
```

```python
def suave2(img):
    img_new = np.copy(img) #cria copia desvinculada da original
    
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            soma = np.copy(img[i][j]) # eu tava passando o ponteiro e o valor de soma estava ligado ao array originial --->  dava tudo errado
            cont = 1                  # ambiguidade em atribuir valor de img[i][j] a soma ou atribuir ponteiro a variavel
            
            #em resumo, faz a mdia dos valores de cada pixel
            #soma  um array de dim 3

            if i >=1:  #verificacao de fronteira
                soma += img[i-1][j] 
                cont += 1
            if i <= (img.shape[0] -2): #verificacao de fronteira
                soma += img[i+1][j] 
                cont += 1
                
            if j >=1: #verificacao de fronteira
                soma += img[i][j-1] 
                cont += 1
                
                #diagonais
                if i >=1:
                    soma += img[i-1][j-1] 
                    cont += 1
                if i <= (img.shape[0] -2):
                    soma += img[i+1][j-1] 
                    cont += 1  
                    
            if j <= (img.shape[1] -2): #verificacao de fronteira
                soma += img[i][j+1] 
                cont += 1
                
                #diagonais
                if i >=1:
                    soma += img[i-1][j+1] 
                    cont += 1
                if i <= (img.shape[0] -2):
                    soma += img[i+1][j+1] 
                    cont += 1  
                    
            img_new[i][j] = soma/cont #atualiza valor suavizado na copia
            
    return img_new
```

```python
N = 10

img_suave = np.copy(img)
for i in range(N): #suavia N vezes
    img_suave = suave1(img_suave)

plt.imshow(img_suave) 
```
