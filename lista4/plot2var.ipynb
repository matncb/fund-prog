```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

f = lambda x,y: np.cos(x)*np.cos(y) #define funo usada
grad_f = lambda x,y: np.array([-np.sin(x)*np.cos(y), -np.cos(x)*np.sin(y)])
```

```python
#gerar os dados para plotar

x_data = np.linspace(-3, 3, 100)
y_data = np.linspace(-3, 3, 100)
x_data, y_data = np.meshgrid(x_data, y_data) #basicamente diz que se trata de uma superficie e no de uma curva

z_data = f(x_data, y_data) #aplica a funo para achar z

#criar a figura e fazer o plot
fig = plt.figure()
ax = plt.axes(projection ='3d')

#ax.view_init(elev=30, azim=45, roll=15) #ponto de vista
#theta, phi, psi --> mais ou menos os angulos de euler

surf = ax.plot_surface(x_data, y_data, z_data, cmap=cm.coolwarm)

fig.colorbar(surf, shrink=0.5, aspect=5) #legenda de cor
```

```python
#plot das curvas de nvel

fig = plt.figure() #gerar figura
ax = plt.axes() #gerar eixo
cs = ax.contourf(x_data, y_data, z_data,  cmap=cm.coolwarm) #curva de nvel
fig.colorbar(cs, ax=ax, shrink=0.9) #legenda de cor

#plot do gradiente

#gerar os dados para plotar
#menos pontos para melhor visualizao
x_data = np.linspace(-3, 3, 15)
y_data = np.linspace(-3, 3, 15)
x_data, y_data = np.meshgrid(x_data, y_data) #basicamente diz que se trata de uma superficie e no de uma curva

z_data = f(x_data, y_data) #aplica a funo para achar z


grad_f_data = grad_f(x_data, y_data)   
ax.quiver(x_data, y_data, grad_f_data[0], grad_f_data[1], scale = 30 )
```
