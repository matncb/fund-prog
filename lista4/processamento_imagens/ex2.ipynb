```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = mpimg.imread('stinkbug.png') #carregar imagem
img.shape
```

```python
def circulo(img, r):
    #escolher um ponto aleatório
    x = np.random.randint(0, img.shape[0])
    y = np.random.randint(0, img.shape[1])
    
    #determina se é branco ou preto
    c = 0
    if np.random.rand() < 0.5: c = 0
    else: c = 1
    
    #todos os pontos que tiverem distnacia menor que r terão seus pixeis trocados
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            raio2 = (i-x)**2 + (j-y)**2
            if raio2< (r**2):
                if c:
                    img[i][j][:] = np.array([0,0,0])
                else:
                    img[i][j][:] = np.array([1,1,1])
```

```python
N= 20 #número de círculos
r = 20 #raio do círculo

for i in range(N):
    circulo(img, 20) #aplica N vezes
    
plt.imshow(img) #mostra o resultado
```
