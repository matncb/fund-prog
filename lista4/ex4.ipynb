```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = mpimg.imread('concrete.jpg') #carregar imagem de teste
plt.imshow(img) 
```

```python
img = img[30:350,100:400,0] #cortar imagem
img = img/np.max(img) #normalizar valores

plt.imshow(img, cmap='gray') 
```

```python
#filtro
limiar = 0.47
filtro = (img > limiar)*1

plt.imshow(filtro, cmap='gray')
```

```python
fracao = np.sum(filtro)/np.size(filtro) #fracao de agregados
fracao
```
