```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n): #varre o array ---> i assume todas as posies
        for j in range(0, n-i-1): #varre todo array desconsiderando as ltimas posies --> para cada i, uma posio no final est correta, por isso (n-i-1)
            if arr[j] > arr[j+1]: #verifica se est desordenado
                arr[j], arr[j+1] = arr[j+1], arr[j] #troca os pares se tiver desordenado
    return arr
```

# Funcionamento do Bubble Sort

Ele funciona comparando pares de elementos adjacentes em uma lista e trocando-os de posio se estiverem fora de ordem. Esse processo  repetido at que a lista esteja completamente ordenada.

---

## Passos do algoritmo
1. **Percorrer a lista:**
   - Comece no primeiro elemento e compare-o com o prximo.
   - Se o elemento atual for maior que o prximo, troque-os de lugar.
   - Continue comparando pares adjacentes ao longo da lista.

2. **"Bolhas" para o final:**
   - Aps uma passagem completa pela lista, o maior elemento "flutua" para a sua posio correta no final da lista.

3. **Repetio:**
   - Repita o processo, ignorando os ltimos elementos j ordenados a cada iterao.
   - Continue at que nenhuma troca seja necessria, indicando que a lista est ordenada.

---

## Visualizao de um exemplo

Considere a lista inicial: **[5, 3, 8, 6, 2]**

### Primeira passagem:
- Compare 5 e 3: Troca  **[3, 5, 8, 6, 2]**
- Compare 5 e 8: Sem troca  **[3, 5, 8, 6, 2]**
- Compare 8 e 6: Troca  **[3, 5, 6, 8, 2]**
- Compare 8 e 2: Troca  **[3, 5, 6, 2, 8]**

### Segunda passagem:
- Compare 3 e 5: Sem troca  **[3, 5, 6, 2, 8]**
- Compare 5 e 6: Sem troca  **[3, 5, 6, 2, 8]**
- Compare 6 e 2: Troca  **[3, 5, 2, 6, 8]**

### Terceira passagem:
- Compare 3 e 5: Sem troca  **[3, 5, 2, 6, 8]**
- Compare 5 e 2: Troca  **[3, 2, 5, 6, 8]**

### Quarta passagem:
- Compare 3 e 2: Troca  **[2, 3, 5, 6, 8]**

Agora a lista est ordenada: **[2, 3, 5, 6, 8]**.

---

## Caractersticas do Bubble Sort
1. **Complexidade de tempo:**
   - **Pior caso**: O(n^2) (quando a lista est em ordem inversa).
   - **Melhor caso**: O(n) (quando a lista j est ordenada, com uma otimizao que detecta isso).

2. **Complexidade de espao:**
   -  O(1), pois o algoritmo funciona "in-place", ou seja, no utiliza espao extra significativo.

3. **Simples, mas ineficiente:**
   -  fcil de implementar e compreender, mas pouco eficiente para listas grandes.



```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n): #varre todos os elementos da lista --> i assume todas as posies
        min_idx = i #indice que divide parte ordenada da parte no ordenada ---> inicialmente 0 pq nada foi ordenado ainda
        for j in range(i+1, n): #varre todo o array, desconsiderando a parte j ordenada
            if arr[j] < arr[min_idx]: #achar menor elemento da parte no ordenada
                min_idx = j #salva posio
        arr[i], arr[min_idx] = arr[min_idx], arr[i] #troca primeiro elemento da parte no ordenada pelo menor elemento da parte no ordenada --> adiciona mais um elemento a parte ordenada
    return arr
```

# Funcionamento do Selection Sort

O **Selection Sort** divide a lista em duas partes: a parte ordenada e a parte no ordenada. Ele encontra o menor elemento na parte no ordenada e o move para a parte ordenada. Esse processo se repete at que toda a lista esteja ordenada.

---

## Passos do algoritmo
1. **Dividir a lista em partes:**
   - Inicialmente, considere que a parte ordenada est vazia e a no ordenada contm todos os elementos.

2. **Encontrar o menor elemento:**
   - Procure o menor elemento na parte no ordenada.

3. **Trocar o elemento:**
   - Troque o menor elemento encontrado com o primeiro elemento da parte no ordenada.

4. **Repetir o processo:**
   - Mova o "limite" da parte ordenada para frente e repita at que toda a lista esteja ordenada.

---

## Visualizao de um exemplo

Considere a lista inicial: **[5, 3, 8, 6, 2]**

### Primeira iterao:
- Menor elemento na parte no ordenada: **2**
- Troque 2 com 5  **[2, 3, 8, 6, 5]**

### Segunda iterao:
- Menor elemento na parte no ordenada: **3**
- Sem troca, pois 3 j est no lugar  **[2, 3, 8, 6, 5]**

### Terceira iterao:
- Menor elemento na parte no ordenada: **5**
- Troque 5 com 8  **[2, 3, 5, 6, 8]**

### Quarta iterao:
- Menor elemento na parte no ordenada: **6**
- Sem troca, pois 6 j est no lugar  **[2, 3, 5, 6, 8]**

Agora a lista est ordenada: **[2, 3, 5, 6, 8]**.

---

## Caractersticas do Selection Sort
1. **Complexidade de tempo:**
   - **Pior caso e melhor caso**: \(O(n^2)\), pois o algoritmo sempre percorre a lista para encontrar o menor elemento, independentemente de sua ordem.

2. **Complexidade de espao:**
   -  \(O(1)\), j que funciona "in-place", sem necessidade de estrutura de dados auxiliar.

3. **Eficincia:**
   - Apesar de simples, o Selection Sort  menos eficiente que outros algoritmos, como o Merge Sort ou Quick Sort, para listas grandes.


```python
def quick_sort(arr):
    if len(arr) <= 1: #condio para a finalizao da recurso ---> se a lista da esquerda ou da direita forem nulas ou tiverem apenas um elemento, retorna esse array
        return arr

    pivot = arr[len(arr) // 2] #escolhe o piv como metade do array

    left = [x for x in arr if x < pivot] #todos os elementos menores que o piv
    middle = [x for x in arr if x == pivot] #todos os elementos iguais ao piv
    right = [x for x in arr if x > pivot] #todos os elementos maiores que o piv

    return quick_sort(left) + middle + quick_sort(right) #retorna, por recurso, a lista ordenada
```


# Funcionamento do Quick Sort

Ele segue a abordagem de **diviso e conquista** para ordenar os elementos, dividindo repetidamente a lista em sublistas menores baseadas em um piv.

---

## Passos do algoritmo

1. **Escolher um piv:**
   - Selecione um elemento da lista como piv (pode ser o primeiro, ltimo, ou qualquer outro elemento).

2. **Particionar a lista:**
   - Reorganize os elementos para que todos os menores que o piv fiquem  esquerda e todos os maiores fiquem  direita.

3. **Aplicar recurso:**
   - Recursivamente aplique o Quick Sort s sublistas da esquerda e da direita (excluindo o piv, que j est na posio correta).

4. **Finalizar:**
   - Quando as sublistas tiverem tamanho 0 ou 1, o algoritmo est concludo.

---

## Visualizao de um exemplo

Considere a lista inicial: **[5, 3, 8, 6, 2]**

### Primeira etapa (escolha do piv):
- Escolha o piv: **5**
- Particione a lista:
  - Menores que 5: **[3, 2]**
  - Piv: **[5]**
  - Maiores que 5: **[8, 6]**
- Lista aps a partio: **[3, 2, 5, 8, 6]**

### Sublista esquerda: [3, 2]
- Escolha o piv: **3**
- Particione:
  - Menores que 3: **[2]**
  - Piv: **[3]**
  - Maiores que 3: **[]**
- Lista ordenada: **[2, 3]**

### Sublista direita: [8, 6]
- Escolha o piv: **8**
- Particione:
  - Menores que 8: **[6]**
  - Piv: **[8]**
  - Maiores que 8: **[]**
- Lista ordenada: **[6, 8]**

### Combinao final:
- Combine as partes ordenadas: **[2, 3] + [5] + [6, 8] = [2, 3, 5, 6, 8]**

Agora a lista est ordenada: **[2, 3, 5, 6, 8]**.

---

## Caractersticas do Quick Sort

1. **Complexidade de tempo:**
   - **Melhor caso**: \(O(n \log n)\), quando o piv divide a lista em partes balanceadas.
   - **Pior caso**: \(O(n^2)\), quando o piv  escolhido de forma inadequada (como em uma lista j ordenada).

2. **Complexidade de espao:**
   - **Em mdia**: \(O(\log n)\), devido  pilha de recurso.

3. **Eficincia:**
   - O Quick Sort  rpido e amplamente utilizado, especialmente em implementaes otimizadas.


```python
import numpy as np
import matplotlib.pyplot as plt
import time
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
N_list = np.array([100,200,400,800,1600])
t1 = []
t2 = []
t3 = []

for N in N_list:
    arr = np.random.randint(10,size = N) #cria array aleatrio

    #aplica em cada algoritmo

    start_time = time.perf_counter()
    bubble_sort(arr)
    end_time = time.perf_counter()
    t1.append(end_time-start_time)

    start_time = time.perf_counter()
    selection_sort(arr)
    end_time = time.perf_counter()
    t2.append(end_time-start_time)

    start_time = time.perf_counter()
    quick_sort(arr)
    end_time = time.perf_counter()
    t3.append(end_time-start_time)

t1 = np.array(t1)
t2 = np.array(t2)
t3 = np.array(t3)
```

```python
plt.scatter(N_list, t1, label = "bubble_sort")
plt.scatter(N_list, t2, label = "selection_sort")
plt.scatter(N_list, t3, label = "quick_sort")
plt.legend()
plt.title("Comparativo entre os mtodos")
plt.xlabel("N")
plt.ylabel("t(s)")
```
