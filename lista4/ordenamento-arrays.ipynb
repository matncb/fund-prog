```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n): #varre o array ---> i assume todas as posições
        for j in range(0, n-i-1): #varre todo array desconsiderando as últimas posições --> para cada i, uma posição no final está correta, por isso (n-i-1)
            if arr[j] > arr[j+1]: #verifica se está desordenado
                arr[j], arr[j+1] = arr[j+1], arr[j] #troca os pares se tiver desordenado
    return arr
```

# Funcionamento do Bubble Sort

Ele funciona comparando pares de elementos adjacentes em uma lista e trocando-os de posição se estiverem fora de ordem. Esse processo é repetido até que a lista esteja completamente ordenada.

---

## Passos do algoritmo
1. **Percorrer a lista:**
   - Comece no primeiro elemento e compare-o com o próximo.
   - Se o elemento atual for maior que o próximo, troque-os de lugar.
   - Continue comparando pares adjacentes ao longo da lista.

2. **"Bolhas" para o final:**
   - Após uma passagem completa pela lista, o maior elemento "flutua" para a sua posição correta no final da lista.

3. **Repetição:**
   - Repita o processo, ignorando os últimos elementos já ordenados a cada iteração.
   - Continue até que nenhuma troca seja necessária, indicando que a lista está ordenada.

---

## Visualização de um exemplo

Considere a lista inicial: **[5, 3, 8, 6, 2]**

### Primeira passagem:
- Compare 5 e 3: Troca → **[3, 5, 8, 6, 2]**
- Compare 5 e 8: Sem troca → **[3, 5, 8, 6, 2]**
- Compare 8 e 6: Troca → **[3, 5, 6, 8, 2]**
- Compare 8 e 2: Troca → **[3, 5, 6, 2, 8]**

### Segunda passagem:
- Compare 3 e 5: Sem troca → **[3, 5, 6, 2, 8]**
- Compare 5 e 6: Sem troca → **[3, 5, 6, 2, 8]**
- Compare 6 e 2: Troca → **[3, 5, 2, 6, 8]**

### Terceira passagem:
- Compare 3 e 5: Sem troca → **[3, 5, 2, 6, 8]**
- Compare 5 e 2: Troca → **[3, 2, 5, 6, 8]**

### Quarta passagem:
- Compare 3 e 2: Troca → **[2, 3, 5, 6, 8]**

Agora a lista está ordenada: **[2, 3, 5, 6, 8]**.

---

## Características do Bubble Sort
1. **Complexidade de tempo:**
   - **Pior caso**: O(n^2) (quando a lista está em ordem inversa).
   - **Melhor caso**: O(n) (quando a lista já está ordenada, com uma otimização que detecta isso).

2. **Complexidade de espaço:**
   - É O(1), pois o algoritmo funciona "in-place", ou seja, não utiliza espaço extra significativo.

3. **Simples, mas ineficiente:**
   - É fácil de implementar e compreender, mas pouco eficiente para listas grandes.



```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n): #varre todos os elementos da lista --> i assume todas as posições
        min_idx = i #indice que divide parte ordenada da parte não ordenada ---> inicialmente 0 pq nada foi ordenado ainda
        for j in range(i+1, n): #varre todo o array, desconsiderando a parte já ordenada
            if arr[j] < arr[min_idx]: #achar menor elemento da parte não ordenada
                min_idx = j #salva posição
        arr[i], arr[min_idx] = arr[min_idx], arr[i] #troca primeiro elemento da parte não ordenada pelo menor elemento da parte não ordenada --> adiciona mais um elemento a parte ordenada
    return arr
```

# Funcionamento do Selection Sort

O **Selection Sort** divide a lista em duas partes: a parte ordenada e a parte não ordenada. Ele encontra o menor elemento na parte não ordenada e o move para a parte ordenada. Esse processo se repete até que toda a lista esteja ordenada.

---

## Passos do algoritmo
1. **Dividir a lista em partes:**
   - Inicialmente, considere que a parte ordenada está vazia e a não ordenada contém todos os elementos.

2. **Encontrar o menor elemento:**
   - Procure o menor elemento na parte não ordenada.

3. **Trocar o elemento:**
   - Troque o menor elemento encontrado com o primeiro elemento da parte não ordenada.

4. **Repetir o processo:**
   - Mova o "limite" da parte ordenada para frente e repita até que toda a lista esteja ordenada.

---

## Visualização de um exemplo

Considere a lista inicial: **[5, 3, 8, 6, 2]**

### Primeira iteração:
- Menor elemento na parte não ordenada: **2**
- Troque 2 com 5 → **[2, 3, 8, 6, 5]**

### Segunda iteração:
- Menor elemento na parte não ordenada: **3**
- Sem troca, pois 3 já está no lugar → **[2, 3, 8, 6, 5]**

### Terceira iteração:
- Menor elemento na parte não ordenada: **5**
- Troque 5 com 8 → **[2, 3, 5, 6, 8]**

### Quarta iteração:
- Menor elemento na parte não ordenada: **6**
- Sem troca, pois 6 já está no lugar → **[2, 3, 5, 6, 8]**

Agora a lista está ordenada: **[2, 3, 5, 6, 8]**.

---

## Características do Selection Sort
1. **Complexidade de tempo:**
   - **Pior caso e melhor caso**: \(O(n^2)\), pois o algoritmo sempre percorre a lista para encontrar o menor elemento, independentemente de sua ordem.

2. **Complexidade de espaço:**
   - É \(O(1)\), já que funciona "in-place", sem necessidade de estrutura de dados auxiliar.

3. **Eficiência:**
   - Apesar de simples, o Selection Sort é menos eficiente que outros algoritmos, como o Merge Sort ou Quick Sort, para listas grandes.


```python
def quick_sort(arr):
    if len(arr) <= 1: #condição para a finalização da recursão ---> se a lista da esquerda ou da direita forem nulas ou tiverem apenas um elemento, retorna esse array
        return arr

    pivot = arr[len(arr) // 2] #escolhe o pivô como metade do array

    left = [x for x in arr if x < pivot] #todos os elementos menores que o pivô
    middle = [x for x in arr if x == pivot] #todos os elementos iguais ao pivô
    right = [x for x in arr if x > pivot] #todos os elementos maiores que o pivô

    return quick_sort(left) + middle + quick_sort(right) #retorna, por recursão, a lista ordenada
```


# Funcionamento do Quick Sort

Ele segue a abordagem de **divisão e conquista** para ordenar os elementos, dividindo repetidamente a lista em sublistas menores baseadas em um pivô.

---

## Passos do algoritmo

1. **Escolher um pivô:**
   - Selecione um elemento da lista como pivô (pode ser o primeiro, último, ou qualquer outro elemento).

2. **Particionar a lista:**
   - Reorganize os elementos para que todos os menores que o pivô fiquem à esquerda e todos os maiores fiquem à direita.

3. **Aplicar recursão:**
   - Recursivamente aplique o Quick Sort às sublistas da esquerda e da direita (excluindo o pivô, que já está na posição correta).

4. **Finalizar:**
   - Quando as sublistas tiverem tamanho 0 ou 1, o algoritmo está concluído.

---

## Visualização de um exemplo

Considere a lista inicial: **[5, 3, 8, 6, 2]**

### Primeira etapa (escolha do pivô):
- Escolha o pivô: **5**
- Particione a lista:
  - Menores que 5: **[3, 2]**
  - Pivô: **[5]**
  - Maiores que 5: **[8, 6]**
- Lista após a partição: **[3, 2, 5, 8, 6]**

### Sublista esquerda: [3, 2]
- Escolha o pivô: **3**
- Particione:
  - Menores que 3: **[2]**
  - Pivô: **[3]**
  - Maiores que 3: **[]**
- Lista ordenada: **[2, 3]**

### Sublista direita: [8, 6]
- Escolha o pivô: **8**
- Particione:
  - Menores que 8: **[6]**
  - Pivô: **[8]**
  - Maiores que 8: **[]**
- Lista ordenada: **[6, 8]**

### Combinação final:
- Combine as partes ordenadas: **[2, 3] + [5] + [6, 8] = [2, 3, 5, 6, 8]**

Agora a lista está ordenada: **[2, 3, 5, 6, 8]**.

---

## Características do Quick Sort

1. **Complexidade de tempo:**
   - **Melhor caso**: \(O(n \log n)\), quando o pivô divide a lista em partes balanceadas.
   - **Pior caso**: \(O(n^2)\), quando o pivô é escolhido de forma inadequada (como em uma lista já ordenada).

2. **Complexidade de espaço:**
   - **Em média**: \(O(\log n)\), devido à pilha de recursão.

3. **Eficiência:**
   - O Quick Sort é rápido e amplamente utilizado, especialmente em implementações otimizadas.


```python
import numpy as np
import matplotlib.pyplot as plt
import time
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
N_list = np.array([100,200,400,800,1600])
t1 = []
t2 = []
t3 = []

for N in N_list:
    arr = np.random.randint(10,size = N) #cria array aleatório

    #aplica em cada algoritmo

    start_time = time.perf_counter()
    bubble_sort(arr)
    end_time = time.perf_counter()
    t1.append(end_time-start_time)

    start_time = time.perf_counter()
    selection_sort(arr)
    end_time = time.perf_counter()
    t2.append(end_time-start_time)

    start_time = time.perf_counter()
    quick_sort(arr)
    end_time = time.perf_counter()
    t3.append(end_time-start_time)

t1 = np.array(t1)
t2 = np.array(t2)
t3 = np.array(t3)
```

```python
plt.scatter(N_list, t1, label = "bubble_sort")
plt.scatter(N_list, t2, label = "selection_sort")
plt.scatter(N_list, t3, label = "quick_sort")
plt.legend()
plt.title("Comparativo entre os métodos")
plt.xlabel("N")
plt.ylabel("t(s)")
```
