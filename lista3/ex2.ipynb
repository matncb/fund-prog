```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
phi = lambda x: x**(-1/2) #define a funo dada
```

```python
def iteracao(x_0, N): #x_0 --> cond inicial # N nmero de iteracoes
    x = [x_0]
    for i in range(N):
        x.append(phi(x[-1]))
    return x   
```

```python
def plot_xk(x_0, N):
    x = iteracao(x_0, N)
    plt.plot(x)
    plt.xlabel("k")
    plt.ylabel("$x_k$")
    plt.title("$x_k \ por \ k$")
```

```python
plot_xk(0.75, 100)
```

```python
def iteracao_plot(x_0, N): #x_0 --> cond inicial # N nmero de iteracoes
    x_data = [x_0, x_0]
    y_data = [x_0, phi(x_0)]
    for i in range(N):
        x_data.append(phi(x_data[-1]))
        y_data.append(y_data[-1])
        
        x_data.append(x_data[-1])
        y_data.append(phi(x_data[-1]))
    return x_data, y_data   
```

```python
def plot_figura(x_0, N):
    x_data, y_data = iteracao_plot(0.75, 100)
    plt.plot(x_data, y_data, color = "orange")
    
    x = np.linspace(0.5,1.2, 100)
    plt.plot(x,phi(x), label = "y = $\phi (x)$", color = "blue")
    plt.plot(x,x, label = "y = x", color = "black")
    plt.xlabel("x")
    plt.ylabel("y")
    
    plt.legend()
    plt.title("Iterao de Ponto Fixo")
```

```python
plot_figura(0.75, 100)
```
