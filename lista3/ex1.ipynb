```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fibonacci(N): #nmero de elementos da sequencia ---> retorna toda a sequencia
    a = [0, 1] #condicoes iniciais
    for i in range(N-2):
        a.append(a[-1] + a[-2]) #calcula prximo elemento
    return a 
```

```python
fibonacci(10) #sequncia at o dcimo nmero
```

```python
def an_fibonaci(n): #retorna elemento n da sequencia de fibonacci --> mtodo recursivo
    if n == 0: return 0
    elif n == 1: return 1
    else: return an_fibonaci(n-1) + an_fibonaci(n-2)
```

```python
an_fibonaci(9) #nono nmero da sequncia
```

```python
#sequencia usando essa funcao ---> muito menos eficiente que usar a outra forma
def sequencia(N):
    a = []
    for i in range(N):
        a.append(an_fibonaci(i))
    return a
```

```python
sequencia(10) #sequncia at o dcimo nmero pelo mtodo recursivo
```
