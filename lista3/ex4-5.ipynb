```python
import numpy as np
```

```python
def solucao(A, B, n, epsilon): #Anxn, Bnx1, epsilon --> erro mximo adimitido 
    X = np.ones(shape=n)
    delta = np.inf
    
    D = np.diag(np.diag(A))
    L = np.tril(A) -D
    U = np.triu(A) -D

    m_inv = np.linalg.inv(D)
    soma = L+U
    
    while delta > epsilon:
        X_new =  m_inv@ (B - soma@X)
        delta = np.max(np.abs(X_new - X))
        X = X_new
        
    return X
```

```python
#exercicio 3 por esse mtodo para ver se funciona

A = np.array([
    [3, -1],
    [-2, 4],
])

n = 2
B = np.array(([2,1]))

solucao(A, B, n, 1e-10)
```

```python
#sistema pedido pelo exerccio 4

A = np.array([
    [3, -1, -1],
    [-1, 3, -1],
    [-1, -1, 3]
])

n = 3
B = np.array(([1,2,3]))

solucao(A, B, n, 1e-10)
```
