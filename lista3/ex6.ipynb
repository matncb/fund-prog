```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
#funcao do enunciado
f = lambda x: x*np.exp(-x**2) - 0.1
dfdx = lambda x: (1-2*x**2)*np.exp(-x**2)

epsilon = 1e-15 #erro máximo definido
```

```python
def solucao(x0, f, dfdx, epsilon):
    #defgime as condições iniciais
    delta = np.inf
    x = x0
    cont = 0
    
    #realiza as iteracoes até o erro ser menor que o erro máximo
    while delta > epsilon:
        x = x - f(x)/dfdx(x)
        delta = np.abs(f(x))
        cont += 1
        
    return x, cont
```

```python
#fazer o plot da função
x = np.linspace(-1, 2, 500)
y = f(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot da função")
```

```python
#achar primeira raiz

x_1, cont_1 = solucao(0.1, f, dfdx, epsilon)

print(f"O valor da raíz é {x_1}")
print(f"O número de iterações é {cont_1}")
print(f"O erro é {f(x_1)}")
```

```python
#achar segunda raiz

x_2, cont_2 = solucao(1.6, f, dfdx, epsilon)

print(f"O valor da raíz é {x_2}")
print(f"O número de iterações é {cont_2}")
print(f"O erro é {f(x_2)}")
```
