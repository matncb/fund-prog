```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#4 vezes integral de 0 at 1 de f(x)

f = lambda x:  np.sqrt(1-x**2)

def find_pi(N):
    # N  u nmero de subinstervalos
    delta_x = 1/N
    x_data = np.linspace(0, 1, N)
    
    integral = np.sum(f(x_data)*delta_x)
    
    return 4*integral
```

```python
N = [1e2, 1e3, 1e4, 1e5, 1e6, 1e7, 1e8]
pi = []

for n in N:
    pi.append(find_pi(int(n)))
    
plt.scatter(N, np.abs(np.array(pi) - np.pi))
plt.xscale("log")
plt.yscale("log")

plt.title("Erro x Nmero de Subintervalos")
plt.xlabel("Nmero de Subintervaloss")
plt.ylabel("Erro")
```
