```python
def caminhada_2d(n_passos, a):
    r_0 = (0,0)
    r = [r_0]

    escolhas = np.random.choice([0,1,2,3], size = n_passos)
    opcoes = np.array([(0,a), (a, 0), (0, -a), (-a, 0)])
    delta_r = opcoes[escolhas]

    for i in delta_r:
        r.append(r[-1] + i)

    r = np.array(r)
    d = lambda i: i[0]**2 + i[1]**2
    distancia = d(r)
    return  distancia

caminhada_2d(10000, 1)
```
