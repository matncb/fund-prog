```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def caminhada_2d(n_passos, a):
    r_0 = (0,0)
    rx = [r_0[0]]
    ry = [r_0[1]]

    escolhas = np.random.choice([0,1,2,3], size = n_passos)
    opcoes = np.array([(0,a), (a, 0), (0, -a), (-a, 0)])

    delta_rx, delta_ry = opcoes[escolhas].T

    for i in delta_rx:
        rx.append(rx[-1] + i)

    for i in delta_ry:
        ry.append(ry[-1] + i)

    return  np.array(rx), np.array(ry)

def plot_caminhada_2d(rx, ry, num_passos,a):
    plt.plot(rx, ry)
    plt.title("{} passos, a = {} ".format(num_passos, a))
    plt.xlabel("x")
    plt.ylabel("y")

def main_2d(num_passos, a, analise = False, M = 1):
    if analise:

        avg_distancia = np.zeros(num_passos+1)
        for i in range(M):
            rx, ry = caminhada_2d(num_passos, a)
            avg_distancia += rx**2 + ry**2
        avg_distancia = avg_distancia/M

        plt.plot(avg_distancia, marker = 'o', ms = 1,  ls = '')
        plt.xlabel("Passos")
        plt.ylabel("Distancia Quadrática Média")
        plt.title("Distância Quadrática Média x Passos")
    else:
        rx, ry = caminhada_2d(num_passos, a)
        plot_caminhada_2d(rx,ry, num_passos,a)

    
```

```python
main_2d(10000,1)
```

```python
main_2d(10000,1, analise=True, M=1000)
```

```python
def caminhada_3d(n_passos, a):
    r_0 = (0,0,0)
    rx = [r_0[0]]
    ry = [r_0[1]]
    rz = [r_0[2]]

    escolhas = np.random.choice([0,1,2,3,4,5], size = n_passos)
    opcoes = np.array([(a,0, 0), (-a, 0, 0), (0,a,0), (0,-a,0),(0,0,a), (0,0,-a) ])

    delta_rx, delta_ry, detlta_rz = opcoes[escolhas].T

    for i in delta_rx:
        rx.append(rx[-1] + i)

    for i in delta_ry:
        ry.append(ry[-1] + i)

    for i in detlta_rz:
        rz.append(rz[-1] + i)

    return  np.array(rx), np.array(ry), np.array(rz)

def plot_caminhada_3d(rx, ry, rz, num_passos,a):

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(rx, ry, rz, s= 0.1)
    plt.title("{} passos, a = {} ".format(num_passos, a))

def main_3d(num_passos, a, analise = False, M = 1):
    if analise:

        avg_distancia = np.zeros(num_passos+1)
        for i in range(M):
            rx, ry, rz = caminhada_3d(num_passos, a)
            avg_distancia += rx**2 + ry**2 + rz**2
        avg_distancia = avg_distancia/M

        plt.plot(avg_distancia, marker = 'o', ms = 1,  ls = '')
        plt.xlabel("Passos")
        plt.ylabel("Distancia Quadrática Média")
        plt.title("Distância Quadrática Média x Passos")
    else:
        rx, ry, rz = caminhada_3d(num_passos, a)
        plot_caminhada_3d(rx,ry,rz, num_passos,a)
```

```python
main_3d(10000,1)
```

```python
main_3d(10000,1, analise=True, M=1000)
```
