```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
#cálculo exato

import sympy as smp

x = smp.Symbol('x')
f = 1 + (1/2)*smp.sin(2*x)**3
g = 3+ (1/2)*smp.cos(3*x)**5

valor = smp.integrate(g-f, (x, 0, 2*smp.pi)).evalf()  #4pi
print("A área é: " + str(valor))
```

```python
f = lambda x: 1 + (1/2)*np.sin(2*x)**3
g = lambda x: 3+ (1/2)*np.cos(3*x)**5

x = np.linspace(0,2*np.pi, int(1e3))

plt.plot(x, f(x), label = 'f')
plt.plot(x, g(x), label = 'g')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráficos de f e g")
plt.legend()
```

```python
#tomemos um retângulo de 0 a 2pi no eixo x e de 0.5 a 3.5 no eixo y ---> A = 2*pi*3

def monte_carlo(num_pontos):
    x = 2*np.pi*np.random.rand(num_pontos)
    y = 3*np.random.rand(num_pontos) + 0.5

    condicao = ((y > f(x))  & (y< g(x)))
    not_condicao =  condicao == 0

    plt.scatter(x[condicao], y[condicao], color = 'red', s = 1)
    plt.scatter(x[not_condicao], y[not_condicao], color = 'blue', s=1)

    return condicao, not_condicao

def area(num_pontos, condicao, not_condicao):
    A = np.sum(np.int16(condicao))/num_pontos * 2*np.pi*3
    print("A área é: " + str(A))
```

```python
num_pontos = int(1e7)

condicao, not_condicao = monte_carlo(num_pontos)
area(num_pontos, condicao, not_condicao)
```
