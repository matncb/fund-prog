```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
#clculo exato

import sympy as smp

x = smp.Symbol('x')
f = 1 + (1/2)*smp.sin(2*x)**3
g = 3+ (1/2)*smp.cos(3*x)**5

valor = smp.integrate(g-f, (x, 0, 2*smp.pi)).evalf()  #4pi
print("A rea : " + str(valor))
```

```python
f = lambda x: 1 + (1/2)*np.sin(2*x)**3
g = lambda x: 3+ (1/2)*np.cos(3*x)**5

x = np.linspace(0,2*np.pi, int(1e3))

plt.plot(x, f(x), label = 'f')
plt.plot(x, g(x), label = 'g')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Grficos de f e g")
plt.legend()
```

```python
#tomemos um retngulo de 0 a 2pi no eixo x e de 0 a 3.5 no eixo y ---> A = 2*pi*3.5
num_pontos = int(1e8)

x = 2*np.pi*np.random.rand(num_pontos)
y = 3.5*np.random.rand(num_pontos)

A = np.sum(np.int16((y > f(x))  & (y< g(x))))/num_pontos * 2*np.pi*3.5
print("A rea : " + str(A))
```
