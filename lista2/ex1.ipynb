```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def histograma(N, M): # M dados jogados N vezes
    matriz = np.random.randint(1,7, size=(int(N),int(M))) # gera uma matriz N x M , numro de lanamentos x nmero de dados
    dados = [] #ter dimenso N                           # cada linha deve ser somada e constiruir uma soma de M dados
    for i in range(int(N)):
        dados.append(np.sum(matriz[i]))
    return dados #retorna os N valores da soma dos M dados
```

```python
def update(frame, ax, N, M):
    dados = histograma(N[frame], M) #gera os dados para fazer o histograma
    ax.clear()

    ax.hist(dados,  bins = np.arange(M, 6*M, 1), histtype='step') #faz o histograma

    ax.set_xlabel("Valor da Soma")
    ax.set_ylabel("Frequncia")
    ax.set_xlim(M, 6*M) #valor mnimo e mximo possvel da soma
    ax.set_title("Soma de {} dados repetida {} vezes".format(M, int(N[frame])))
```

```python
def gerar_gif(M):
    N = [1e2, 1e3, 1e4, 1e5, 1e6] # Valores que cada um dos M dados sero jogados
    dados = histograma(N[0], M)

    fig, ax = plt.subplots()
    ax.hist(dados, bins = np.arange(M, 6*M, 1), histtype='step')

    ax.set_xlabel("Valor da Soma")
    ax.set_ylabel("Frequncia")
    ax.set_xlim(M, 6*M)
    ax.set_title("Soma de {} dados repetida {} vezes".format(M,int(N[0])))
    
    
    ani = animation.FuncAnimation(fig, update, fargs=(ax, N, M), frames = len(N), interval = 1500)  #Faz a animao
    ani.save(filename= "./gifs_ex1/ex1-{}-dados.gif".format(M), writer="pillow") #Salva a animao
```

```python
M_valores = [2,4,6,8,10]

for M in M_valores:  #Faz animao para todos os valores de M pedidos
    gerar_gif(M) 
```
