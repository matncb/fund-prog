```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def histograma(N, M):
    matriz = np.random.randint(1,7, size=(int(N),int(M)))
    dados = []
    for i in range(int(N)):
        dados.append(np.sum(matriz[i]))
    return dados
```

```python
num_div = 10
```

```python
def update(frame, ax, N, M):
    dados = histograma(N[frame], M)
    ax.clear()
    ax.hist(dados, num_div)
    ax.set_xlabel("Valor da Soma")
    ax.set_ylabel("Frequncia")
    ax.set_xlim(M, 6*M)
    ax.set_title("Soma de {} dados repetida {} vezes".format(M, int(N[frame])))
```

```python
def gerar_gif(M):
    N = [1e2, 1e3, 1e4, 1e5, 1e6]
    dados = histograma(N[0], M)

    fig, ax = plt.subplots()
    ax.hist(dados, num_div)
    
    ax.set_xlabel("Valor da Soma")
    ax.set_ylabel("Frequncia")
    ax.set_xlim(M, 6*M)
    ax.set_title("Soma de {} dados repetida {} vezes".format(M,int(N[0])))
    
    
    ani = animation.FuncAnimation(fig, update, fargs=(ax, N, M), frames = len(N), interval = 500)
    ani.save(filename= "ex1.gif", writer="pillow")
```

```python
gerar_gif(10)
```
