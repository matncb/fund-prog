```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
f = lambda x: 1 + (1/2)*np.sin(2*x)**3
g = lambda x: 3+ (1/2)*np.cos(3*x)**5

x = np.linspace(0,2*np.pi, int(1e3))

p1 = lambda y: (1/1.5625)*(y-1)*(y-3.5)
p2 = lambda y : -(1/1.5625)*(y-1)*(y-3.5) + 2*np.pi

y = np.linspace(1,3.5, int(1e3))
plt.plot(p1(y), y, label = 'p1')
plt.plot(p2(y), y,label = 'p2')


plt.plot(x, f(x), label = 'f')
plt.plot(x, g(x), label = 'g')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Grfico da figura")
```

```python
#tomemos um retngulo de -1 a 2pi+1 no eixo x e de 0 a 3 no eixo y ---> A = (2*pi+2)*3

def monte_carlo(num_pontos):
    x = (2*np.pi +2)*np.random.rand(num_pontos) -1 
    y = 3*np.random.rand(num_pontos) + 0.5

    condicao = ((x>0) & (x<(2*np.pi)) & (y > f(x))  & (y< g(x))) | ((x<0) & (x>p1(y))) | ((x>(2*np.pi)) & (x<p2(y)))
    not_condicao =  condicao == 0

    plt.scatter(x[condicao], y[condicao], color = 'red', s = 1)
    plt.scatter(x[not_condicao], y[not_condicao], color = 'blue', s=1)

    return condicao, not_condicao

def area(num_pontos, condicao,  not_condicao):
    A = np.sum(np.int16(condicao))/num_pontos * (2*np.pi+2)*3
    print("A rea : " + str(A))
```

```python
num_pontos = int(1e7)

condicao, not_condicao = monte_carlo(num_pontos)
area(num_pontos, condicao, not_condicao)
```
