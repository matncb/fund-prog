```python
import numpy as np
import time

import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
#declarar função

low = -1
high = 1

def multiplica_matriz(m, r, n):
    A = np.random.randint(low, high, size = (int(m),int(r)))
    B = np.random.randint(low, high, size = (int(r),int(n)))

    start_time = time.perf_counter()
    C = A@B
    end_time = time.perf_counter()

    execution_time = end_time - start_time
    return execution_time, C
```

```python
#carregar dados

x_data = [1000, 2000, 3000, 4000, 5000]
y_data = []

for dimensao in x_data:
    print("Calculando... Dimensão = " + str(dimensao))
    tempo, C = multiplica_matriz(dimensao, dimensao, dimensao)
    y_data.append(tempo)
    print("Feito. Dimensão = " + str(dimensao))
    
y_data = np.array(y_data)
```

```python
#plot do gráfico em escala normal

plt.scatter(x_data, y_data)

plt.title("Tempo de execução x Dimensão")
plt.xlabel("Dimensão")
plt.ylabel("Tempo de Execução (s)")
```

```python
#plot do gráfico em escala log-log

plt.scatter(x_data, y_data)

plt.title("Tempo de execução x Dimensão")
plt.xlabel("Dimensão")
plt.ylabel("Tempo de Execução (s)")

plt.yscale('log')
plt.xscale('log')
```
