# Primeira Função

$c = \alpha a + \beta b$

numpy.random.randint(low, high=None, size=None, dtype=’l’) – gera uma matriz de inteiros aleatórios entre low e high, excluindo high.

```python
low = -1
high = 1

def tempo_vetor(n):
    alpha = np.random.randint(low, high, size = 1)[0]
    beta = np.random.randint(low, high, size = 1)[0]

    a = np.random.randint(low, high, size = int(n))
    b = np.random.randint(low, high, size = int(n))
    
    start_time = time.perf_counter()
    c = alpha*a + beta*b
    end_time = time.perf_counter()

    execution_time = end_time - start_time
    return execution_time

def dados(x_data, N):
    y_data = []
    y_data_err = []


    for n in x_data:
        l = []
        for i in range(N):
            l.append(tempo_vetor(n))
        l = np.array(l)
        y_data.append(l.mean())
        y_data_err.append(l.std())

    return y_data, y_data_err
```

```python
#carregar dados

x_data = np.linspace(1, 100000000, 20)
y_data, y_data_err = dados(x_data, 3)
```

```python
#plot em escala normal

plt.errorbar(x_data,y_data, yerr=y_data_err, marker = 'o', ls = '')

plt.title("Tempo de execução x Dimensão")
plt.xlabel("Dimensão")
plt.ylabel("Tempo de Execução (s)")
```

```python
#plot em escala log-log

plt.errorbar(x_data,y_data, yerr=y_data_err, marker = 'o', ls = '')

plt.title("Tempo de execução x Dimensão")
plt.xlabel("Dimensão")
plt.ylabel("Tempo de Execução (s)")

plt.yscale('log')
plt.xscale('log')
```

# Segunda Função

```python
low = -1
high = 1

def tempo_matriz(n, m):
    A = np.random.randint(low, high, size = (n,n))

    start_time = time.perf_counter()
    Am = np.linalg.matrix_power(A, m)
    end_time = time.perf_counter()

    execution_time = end_time - start_time
    return execution_time
```

```python
#gerar dados

x_data = np.array([1000, 2000, 3000])
m_data = np.array([2,3])

y_data_array = []

for m in m_data:
    y_data = []
    for n in x_data:
        y_data.append(tempo_matriz(n,m))
    y_data_array.append(y_data) 
```

```python
#plotar dados em escala normal

m = 2
for y_data in y_data_array:
    plt.scatter(x_data, y_data, label = 'm = ' + str(m))
    m += 1

plt.title("Tempo de execução x Dimensão")
plt.xlabel("Dimensão")
plt.ylabel("Tempo de Execução (s)")
    
plt.legend()
```

```python
#plotar dados em escala log-log

m = 2
for y_data in y_data_array:
    plt.scatter(x_data, y_data, label = 'm = ' + str(m))
    m += 1

plt.title("Tempo de execução x Dimensão")
plt.xlabel("Dimensão")
plt.ylabel("Tempo de Execução (s)")

plt.yscale('log')
plt.xscale('log')    
     
plt.legend()
```
