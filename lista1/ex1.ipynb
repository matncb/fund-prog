# Primeira Função

$c = \alpha a + \beta b$

numpy.random.randint(low, high=None, size=None, dtype=’l’) – gera uma matriz de inteiros aleatórios entre low e high, excluindo high.

```python
low = -100
high = 100

def tempo_vetor(n):
    alpha = np.random.randint(low, high, size = 1)[0]
    beta = np.random.randint(low, high, size = 1)[0]

    a = np.random.randint(low, high, size = int(n))
    b = np.random.randint(low, high, size = int(n))
    
    start_time = time.perf_counter()
    c = alpha*a + beta*b
    end_time = time.perf_counter()

    execution_time = end_time - start_time
    return execution_time

def dados(x_data, N):
    y_data = []
    y_data_err = []


    for n in x_data:
        l = []
        for i in range(N):
            l.append(tempo_vetor(n))
        l = np.array(l)
        y_data.append(l.mean())
        y_data_err.append(l.std())

    return y_data, y_data_err
```

```python
x_data = np.linspace(1, 100000000, 20)
y_data, y_data_err = dados(x_data, 3)

plt.errorbar(x_data,y_data, yerr=y_data_err, marker = 'o', ls = '')
```

```python
x_data = np.array([int(10**1), int(10**2), int(10**3), int(10**4), int(10**5), int(10**6), int(10**7), int(10**8)])
y_data, y_data_err = dados(x_data, 5)

plt.errorbar(x_data,y_data, yerr=y_data_err, marker = 'o', ls = '')

plt.yscale('log')
plt.xscale('log')
```

```python
low = -100
high = 100

def tempo_matriz(n, m):
    A = np.random.randint(low, high, size = (n,n))

    start_time = time.perf_counter()
    Am = np.linalg.matrix_power(A, m)
    end_time = time.perf_counter()

    execution_time = end_time - start_time
    return execution_time
```
