```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_0 = 0.1
N = 5000

def gerar_sequencia(a):
    sequencia = [x_0]
    
    for n in range(1,N+1):
        x_n = a*sequencia[n-1]*(1-sequencia[n-1])
        sequencia.append(x_n)
        
    return sequencia
```

```python
#valores de a entre 0 e 4, como pedido no enunciado

a_valores = np.linspace(0.1, 4, 5)
x_data = np.array(range(N+1))
y_data_array = []

#gear sequência e calcular média e variância 

for a in a_valores:
    sequencia = gerar_sequencia(a)
    y_data_array.append(sequencia)
    
    np.array(sequencia).var()
    
    x_bar_np = np.array(sequencia).mean()
    x_bar = (1/N) * sum(sequencia)
    
    x_var_np = np.array(sequencia).var(ddof=1)
    x_var = (1/(N-1))* sum([(x_i - x_bar)**2 for x_i in sequencia])
    
    print("-----------------------")
    print("a = " + str(a))
    
    print("Média Numpy:  " + str(x_bar_np))
    print("Média:  " + str(x_bar))
    
    print()
    
    print("Variância Numpy:  " + str(x_var_np))
    print("Variância:  " + str(x_var))
    
    print("-----------------------")
```

```python
#plots das sequências para vários valores de a

for i in range(len(a_valores)):
    plt.scatter(x_data, y_data_array[i], label = "a = " + str(a_valores[i]), s =5)
    
plt.xlabel("n")
plt.ylabel(r"$x_n$")
plt.title(r"$X_n$ por $n$")

plt.legend()
```
