```python
import numpy as np 

import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# Grid size
N = 100

# Create an initial random grid
#p0, p1 = 0.8, 0.2
p0, p1 = 0.1, 0.9
grid = np.random.choice([0, 1], N*N, p=[p0, p1]).reshape(N, N)
```

```python
def update(frameNum, img, grid):
    for i in range(N):
        for j in range(N):
            vizinhos = [(i-1, j-1), (i-1, j),(i-1, j+1), (i, j-1), (i, j+1), (i+1, j-1),(i+1, j), (i+1, j+1) ]
            cont = 0
            for k in vizinhos:
                a, b = k
                if a <= (N-1) and a>=0 and b<=(N-1) and b>=0:
                    if grid[a][b]:
                        cont +=1
            if grid[i][j]:
                if cont <2:
                    grid[i][j] = 0
                if cont > 3:
                    grid[i][j] = 0
            elif cont ==3:
                grid[i][j] = 1
    
    img.set_data(grid)
```

```python
fig, ax = plt.subplots()
img = ax.imshow(grid, interpolation='nearest')
ani = animation.FuncAnimation(fig, update, fargs=(img,grid), frames = 200)


ani.save(filename="vida.gif", writer="pillow")
```
