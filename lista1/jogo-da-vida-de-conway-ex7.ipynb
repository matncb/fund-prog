```python
import numpy as np 

import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def update(frameNum, img, grid, N):
    ref_grid = np.copy(grid)
    for i in range(N):
        for j in range(N):
            vizinhos = [(i-1, j-1), (i-1, j),(i-1, j+1), (i, j-1), (i, j+1), (i+1, j-1),(i+1, j), (i+1, j+1) ]
            cont = 0
            for k in vizinhos:
                a, b = k
                if a <= (N-1) and a>=0 and b<=(N-1) and b>=0:
                    if ref_grid[a][b]:
                        cont +=1
            if ref_grid[i][j]:
                if cont <2:
                    grid[i][j] = 0
                if cont > 3:
                    grid[i][j] = 0
            elif cont ==3:
                grid[i][j] = 1
    
    img.set_data(grid)
```

```python
def gerar_gif(grid,N, nome):

    fig, ax = plt.subplots()
    img = ax.imshow(grid, interpolation='nearest')
    ani = animation.FuncAnimation(fig, update, fargs=(img,grid, N), frames = 200)

    ani.save(filename= "./gifs-conway/conway-" + str(nome) + ".gif", writer="pillow")
    #conway-N-po-p1

def gerar_grid_aleatoria(N, p0, p1):  
    #N ----> tamanho do grid
    
    grid = np.random.choice([0, 1], N*N, p=[p0, p1]).reshape(N, N)
    nome = str(N) + "-"  + str(p0) + "-" + str(p1)
    
    return grid, nome

def gerar_grid(N, vivos):
    grid = np.zeros(shape = (N, N))
    
    for pos in vivos:
        i, j = pos
        grid[i][j] = 1
    
    return grid
```

```python
#grid aleatória
N = 100
grid, nome = gerar_grid_aleatoria(N, 0.2, 0.8)
gerar_gif(grid, N, nome)
```

```python
#still lifes

N = 4
vivos = [(1,1), (2,1), (1,2), (2,2)]
nome = "block"

gerar_gif(gerar_grid(N, vivos), N, nome)
```
