```python
import numpy as np
import matplotlib.pyplot as plt

import scienceplots
plt.style.use(['science', 'notebook', 'grid'])
```

```python
#função para gerar sequência

x_0 = 0.1
N = 5000

def gerar_sequencia(a):
    sequencia = [x_0]
    
    for n in range(1,N+1):
        x_n = a*sequencia[n-1]*(1-sequencia[n-1])
        sequencia.append(x_n)
        
    return sequencia
```

```python
a_valores = np.linspace(0, 4, 500) #x_data

for i in range(len(a_valores)): 
    sequencia = gerar_sequencia(a_valores[i])
    plt.scatter([a_valores[i] for k in range(N+1)][500:], sequencia[500:], s =1 ,marker='.' ,color = 'black')

    # o íncio da sequência é instável
    #pegar depois dos primeiros 500 elementos

plt.title("Diagrama de Bifurcação")
plt.xlabel("a")
plt.ylabel("x_n")
```
