```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class grafo:
    def __init__(self, matriz_arestas, array_coordenadas):
        #matriz arestas ---> n_nos x n_nos
        #array coordenadas ---> n_nos , contm tuplas
        
        self.matriz_arestas = matriz_arestas
        self.array_coordenadas = array_coordenadas
        
    def plot(self):
        tamanho = len(self.array_coordenadas)
        printados = []
        
        plt.axis('off')
        delta_x = 0.03
        delta_y = 0.005
        
        for i in range(tamanho):
            for j in range(tamanho):
                if  self.matriz_arestas[i][j]:
                    x1, y1 = self.array_coordenadas[i]
                    x2, y2 = self.array_coordenadas[j]
                    plt.plot([x1, x2], [y1, y2], marker = 'o', ms = 10, color = 'black')
                    
                    if i not in printados:
                        plt.text(x1 - delta_x, y1 + delta_y, str(i+1), color="red", fontsize=12)
                        printados.append(i)
                    if j not in printados:
                        plt.text(x2 - delta_x, y2 + delta_y, str(j+1), color="red", fontsize=12)
                        printados.append(j)
                                                
    def deletar_no(self, k):
        self.matriz_arestas[k] = [0 for j in range(len(self.array_coordenadas))]
        for i in range(len(self.array_coordenadas)):
            self.matriz_arestas[i][k] = 0
            
    def deletar_arresta(self, i, j):
        self.matriz_arestas[i][j] = 0
        self.matriz_arestas[j][i] = 0
        
    def inserir_arresta(self, i, j):
        self.matriz_arestas[i][j] = 1
        self.matriz_arestas[j][i] = 1
        
```

```python
matriz_arestas = np.zeros(shape=(4, 4))    #matriz simtrica, 4 ns

matriz_arestas[0][1] = 1
matriz_arestas[1][2] = 1
matriz_arestas[2][3] = 1
matriz_arestas[1][3] = 1

matriz_arestas[1][0] = 1
matriz_arestas[2][1] = 1
matriz_arestas[3][2] = 1
matriz_arestas[3][1] = 1

array_coordenadas = np.array([(1,2), (2,4), (3,1), (4,3)])

teste_grafo = grafo(matriz_arestas, array_coordenadas)
teste_grafo.plot()
```

```python
teste_grafo.inserir_arresta(0, 2)
teste_grafo.plot()
```

```python
teste_grafo.deletar_no(0)
teste_grafo.plot()
```

```python
teste_grafo.deletar_arresta(1, 2)
teste_grafo.plot()
```
